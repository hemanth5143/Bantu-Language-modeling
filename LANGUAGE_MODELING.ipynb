```python
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction

# Reference sentence (one of the two sentences)
reference = "maabaho msenga wa kuulanga wekaidi kaitowa mhalati yake na chinhu fana mulima mkulu woukwaka moto chasigwa muibahali selusi ya bahali igaluka damu"

# Candidate sentence (the other of the two sentences)
candidate = "yesu kawalongela, \"nowalongelani kweli, siyo musa yawinileni ndiya ija ya kuulanga maabaho nhaguso na yelile yawo ni ija wa ku\""

# Convert reference to a list of lists (required format for NLTK)
references = [reference.split()]

# Create a SmoothingFunction
smoothing_function = SmoothingFunction().method1

# Calculate BLEU score with smoothing
bleu_score = sentence_bleu(references, candidate.split(), smoothing_function=smoothing_function)

# Print the BLEU score
print("BLEU Score:", bleu_score)
```
